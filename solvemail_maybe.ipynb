**Auth comparison — this is the key question:**

| | Your `google_auth.py` | solvemail `auth.py` |
|---|---|---|
| OAuth (solveit) | ✅ Manual browser flow via pickle | ✅ `oauth_creds()` — JSON token, handles headless/notebook |
| Service account (GH Actions) | ✅ `GOOGLE_SERVICE_ACCOUNT_KEY` env var → `.with_subject()` | ✅ `svc_acct_creds(sa_path, subject=...)` |

solvemail's `svc_acct_creds` reads from a **file path**, not an env var directly, but that's trivial — you'd just write the env var JSON to a temp file, or pass `service_account.Credentials.from_service_account_info(...)` and hand the creds directly to `Gmail(creds=...)`.
